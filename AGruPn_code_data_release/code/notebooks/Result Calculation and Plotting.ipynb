```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.stats import wilcoxon


# 设置中文显示
plt.rcParams["font.family"] = ["Heiti TC"]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
sns.set_style("whitegrid")
sns.set_palette("husl")

# 定义评估函数
def calculate_smape(y_true, y_pred):
    """计算对称平均绝对百分比误差"""
    return 100 * np.mean(np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred) + 1e-8))

def calculate_icr(y_true, lower, upper):
    """计算区间覆盖率"""
    return 100 * np.mean((y_true >= lower) & (y_true <= upper))

def calculate_is(y_true, lower, upper, alpha=0.05):
    """计算区间得分"""
    interval_width = upper - lower
    lower_penalty = (2 / alpha) * (lower - y_true) * (y_true < lower)
    upper_penalty = (2 / alpha) * (y_true - upper) * (y_true > upper)
    return np.mean(interval_width + lower_penalty + upper_penalty)

def calculate_rmse(y_true, y_pred):
    """计算均方根误差"""
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

# 计算所有指标
def calculate_all_metrics(df, dataset_name):
    """计算数据集的所有评估指标"""
    y_true = df['true_value'].values
    y_pred = df['pred_med'].values
    lower = df['pred_low'].values
    upper = df['pred_high'].values
    
    metrics = {
        '数据集': dataset_name,
        '样本数': len(df),
        'SMAPE': calculate_smape(y_true, y_pred),
        'ICR': calculate_icr(y_true, lower, upper),
        'IS': calculate_is(y_true, lower, upper),
    }
    
    return metrics
```

```python
# 读取数据文件
df_0 = pd.read_csv('quantile_prediction_results_Aventa_9_0.csv')
df_1 = pd.read_csv('quantile_prediction_results_Aventa_9_1.csv')
df_all = pd.read_csv('quantile_prediction_results_Aventa_9_all.csv')
 
# 确保列名一致
expected_cols = ['time_index', 'true_value', 'pred_low', 'pred_med', 'pred_high']
for df, name in zip([df_0, df_1, df_all], ['0', '1', 'all']):
    missing_cols = set(expected_cols) - set(df.columns)
    if missing_cols:
        print(f"警告: 文件{name}缺少列: {missing_cols}")

# 合并0和1文件，按time_index排序
df_combined = pd.concat([df_0, df_1], ignore_index=True)
df_combined = df_combined.sort_values('time_index').reset_index(drop=True)
print(calculate_all_metrics(df_all, 'Aventa-0'))
print(calculate_all_metrics(df_combined, 'Aventa-1'))

# Wilcoxon符号秩检验（适用于两个相关样本）
statistic_w, p_value_w = wilcoxon(df_all.true_value, df_all.pred_med)

print("\nWilcoxon符号秩检验结果:")
print(f"统计量: {statistic_w}")
print(f"P值: {p_value_w:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import matplotlib
 # 分别过滤两个数据集（真实值>=0.1）
df_all_filtered = df_all[df_all['true_value'] >= 0.1].copy()
df_combined_filtered = df_combined[df_combined['true_value'] >= 0.1].copy()

# 分别计算两个数据集的预测误差
df_all_filtered['error'] = df_all_filtered['pred_med'] - df_all_filtered['true_value']
df_all_filtered['abs_error'] = np.abs(df_all_filtered['error'])
df_all_filtered['rel_error'] = np.abs(df_all_filtered['error'] / df_all_filtered['true_value']) * 100

df_combined_filtered['error'] = df_combined_filtered['pred_med'] - df_combined_filtered['true_value']
df_combined_filtered['abs_error'] = np.abs(df_combined_filtered['error'])
df_combined_filtered['rel_error'] = np.abs(df_combined_filtered['error'] / df_combined_filtered['true_value']) * 100
# Set scientific publication-style plotting parameters with larger fonts
plt.rcParams.update({
    'font.size': 20,  # Increased from 11
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'axes.labelsize': 20,  # Increased from 12
    'axes.titlesize': 20,  # Increased from 14
    'axes.titleweight': 'bold',
    'axes.linewidth': 4,  # Slightly increased
    'xtick.labelsize': 20,  # Increased from 11
    'ytick.labelsize': 20,  # Increased from 11
    'legend.fontsize': 20,  # Increased from 11
    'legend.frameon': True,
    'legend.framealpha': 0.8,
    'figure.titlesize': 20,  # Increased from 16
    'figure.titleweight': 'bold',
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
    'savefig.pad_inches': 0.1
})

# Function to calculate statistical metrics
def calculate_stats(true_vals, pred_vals, errors):
    """Calculate various statistical metrics"""
    # R²
    r2 = stats.linregress(true_vals, pred_vals).rvalue**2
    # Pearson correlation coefficient
    corr_coef = np.corrcoef(true_vals, pred_vals)[0, 1]
    # MAE
    mae = np.mean(np.abs(errors))
    # RMSE
    rmse = np.sqrt(np.mean(errors**2))
    # MAPE
    mape = np.mean(np.abs(errors / true_vals)) * 100
    
    return {
        'R²': r2,
        'r': corr_coef,
        'MAE': mae,
        'RMSE': rmse,
        'MAPE': mape
    }

# Calculate statistical metrics for both datasets separately
stats_all = calculate_stats(
    df_all_filtered['true_value'].values,
    df_all_filtered['pred_med'].values,
    df_all_filtered['error'].values
)

stats_combined = calculate_stats(
    df_combined_filtered['true_value'].values,
    df_combined_filtered['pred_med'].values,
    df_combined_filtered['error'].values
)

# Create figure
fig = plt.figure(figsize=(8, 16))  # Increased width slightly

# Use Set2 color scheme
colors = plt.cm.Set2(np.linspace(0.3, 0.7, 2))

# Subplot 1: Scatter plot of true values vs predicted values
ax1 = plt.subplot(2, 1, 1)

# Create scatter plots with different markers and transparency
scatter1 = ax1.scatter(df_all_filtered['true_value'], df_all_filtered['pred_med'], 
                      alpha=1, s=30, label='Global Model',  # Increased marker size
                      color=colors[0], edgecolor='white', linewidth=0.8,  # Increased linewidth
                      marker='o')

scatter2 = ax1.scatter(df_combined_filtered['true_value'], df_combined_filtered['pred_med'], 
                      alpha=0.3, s=30, label='Regional Model',  # Increased marker size
                      color=colors[1], edgecolor='white', linewidth=0.8,  # Increased linewidth
                      marker='s')

# Add perfect prediction line
min_val = min(df_all_filtered['true_value'].min(), df_all_filtered['pred_med'].min(),
              df_combined_filtered['true_value'].min(), df_combined_filtered['pred_med'].min())
max_val = max(df_all_filtered['true_value'].max(), df_all_filtered['pred_med'].max(),
              df_combined_filtered['true_value'].max(), df_combined_filtered['pred_med'].max())

ax1.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.8, 
         linewidth=2.0, label='Perfect Prediction Line (y=x)')  # Increased linewidth

# Add regression lines
x_range = np.linspace(min_val, max_val, 100)

# Global model regression
coeff_all = np.polyfit(df_all_filtered['true_value'], df_all_filtered['pred_med'], 1)
y_fit_all = np.polyval(coeff_all, x_range)
ax1.plot(x_range, y_fit_all, color=colors[0], alpha=0.9, linewidth=2.5,  # Increased linewidth
         linestyle='-', label=f'Global Fit: y={coeff_all[0]:.3f}x+{coeff_all[1]:.3f}')

# Regional model regression
coeff_comb = np.polyfit(df_combined_filtered['true_value'], df_combined_filtered['pred_med'], 1)
y_fit_comb = np.polyval(coeff_comb, x_range)
ax1.plot(x_range, y_fit_comb, color=colors[1], alpha=0.9, linewidth=2.5,  # Increased linewidth
         linestyle='-', label=f'Regional Fit: y={coeff_comb[0]:.3f}x+{coeff_comb[1]:.3f}')

ax1.set_xlabel('Measured Value', fontsize=20)  # Increased font size
ax1.set_ylabel('Predicted Value', fontsize=20)  # Increased font size

ax1.legend(loc='lower right', fontsize=15, framealpha=0.9)  # Increased font size
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_aspect('auto')

# Subplot 2: Relative error distribution histogram
ax2 = plt.subplot(2, 1, 2)

# Calculate appropriate bins
max_error = max(df_all_filtered['rel_error'].max(), df_combined_filtered['rel_error'].max())
bins = np.linspace(0, min(max_error, 100), 50)  # 20 bins

# Plot histograms
n_all, bins_all, patches_all = ax2.hist(df_all_filtered['rel_error'], bins=bins, alpha=1, 
                                         label='Global Model', color=colors[0], density=True,
                                         edgecolor='white', linewidth=1.0)  # Increased linewidth

n_comb, bins_comb, patches_comb = ax2.hist(df_combined_filtered['rel_error'], bins=bins, alpha=0.3, 
                                            label='Regional Model', color=colors[1], density=True,
                                            edgecolor='white', linewidth=1.0)  # Increased linewidth

# Add mean lines
mean_all = df_all_filtered['rel_error'].mean()
mean_comb = df_combined_filtered['rel_error'].mean()

ax2.axvline(x=mean_all, color=colors[0], linestyle='--', linewidth=2.5, alpha=0.9,  # Increased linewidth
            label=f'Global Mean: {mean_all:.1f}%')
ax2.axvline(x=mean_comb, color=colors[1], linestyle='--', linewidth=2.5, alpha=0.9,  # Increased linewidth
            label=f'Regional Mean: {mean_comb:.1f}%')

ax2.set_xlabel('Relative Error (%)', fontsize=20)  # Increased font size
ax2.set_ylabel('Probability Density', fontsize=20)  # Increased font size

ax2.legend(loc='upper right', fontsize=15, framealpha=0.9)  # Increased font size
ax2.grid(True, alpha=0.3, linestyle='--')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cm import get_cmap

# 设置图形和字体
plt.rcParams.update({
    'font.size': 20,
    'font.family': 'serif',
    'axes.labelsize': 20,
    'axes.titlesize': 20,
    'axes.titleweight': 'bold',
    'figure.dpi': 300
})

# 创建图形
fig, ax = plt.subplots(figsize=(10, 8))

# 准备数据
data1 = df_all_filtered['abs_error'].values
data2 = df_combined_filtered['abs_error'].values

# 计算累积分布函数 (CDF)
def calculate_cdf(data):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    return sorted_data, cdf

sorted_data1, cdf1 = calculate_cdf(data1)
sorted_data2, cdf2 = calculate_cdf(data2)

# 计算合适的x轴范围（显示90%的数据点）
x_limit = max(np.percentile(sorted_data1, 90), np.percentile(sorted_data2, 90)) * 1.3

# 使用散点图创建渐变效果
cmap_blue = get_cmap('Blues')
cmap_orange = get_cmap('Oranges')

# 绘制全局模型CDF（散点图渐变）
scatter1 = ax.scatter(sorted_data1, cdf1, 
                     c=cdf1,  # 使用CDF值作为颜色映射
                     cmap=cmap_blue,
                     s=8,  # 增加点的大小
                     alpha=0.9,
                     label=f'Global Model (n={len(data1):,})',
                     edgecolors='none')

# 绘制分区模型CDF（散点图渐变）
scatter2 = ax.scatter(sorted_data2, cdf2, 
                     c=cdf2,  # 使用CDF值作为颜色映射
                     cmap=cmap_orange,
                     s=8,  # 增加点的大小
                     alpha=0.9,
                     label=f'Regional Model (n={len(data2):,})',
                     edgecolors='none')

# 添加连接线增强可视化效果
ax.plot(sorted_data1, cdf1, color='#1E88E5', alpha=0.3, linewidth=3)
ax.plot(sorted_data2, cdf2, color='#FF9800', alpha=0.3, linewidth=3)

# 设置标签和标题
ax.set_xlabel('Absolute Error', fontsize=20)
ax.set_ylabel('Cumulative Probability', fontsize=20)
ax.set_title('Cumulative Distribution of Absolute Errors', fontsize=20, fontweight='bold', pad=20)

# 设置坐标轴范围，减少空白
ax.set_xlim(0, x_limit)  # 限制x轴范围
ax.set_ylim(0, 1.05)     # y轴从0到1.05，留一点边距

# 添加网格
ax.grid(True, alpha=0.3, linestyle='--')

# 在曲线上添加关键点标记
key_percentiles = [20,30,40, 50,60, 70,80, 90]
for p in key_percentiles:
    # 全局模型
    perc1 = np.percentile(data1, p)
    cdf_val1 = np.interp(perc1, sorted_data1, cdf1)
    ax.plot(perc1, cdf_val1, 'o', markersize=15, 
            color='blue', markeredgecolor='white', markeredgewidth=1, zorder=3)
    
    # 分区模型
    perc2 = np.percentile(data2, p)
    cdf_val2 = np.interp(perc2, sorted_data2, cdf2)
    ax.plot(perc2, cdf_val2, 's', markersize=15, 
            color='orange', markeredgecolor='white', markeredgewidth=1, zorder=3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import cm
from matplotlib.colors import Normalize

# 设置图形和字体
plt.rcParams.update({
    'font.size': 14,
    'font.family': 'sans-serif',
    'axes.labelsize': 16,
    'axes.titlesize': 18,
    'axes.titleweight': 'bold',
    'figure.dpi': 300,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'grid.linestyle': '--'
})

# 原始数据
data = {
    'Global Model': {
        'Site1': {'SMAPE': 3.576, 'ICR': 90.035, 'IS': 3.800},
        'Site2': {'SMAPE': 3.422, 'ICR': 91.084, 'IS': 4.300},
        'Site3': {'SMAPE': 3.278, 'ICR': 89.336, 'IS': 4.101},
        'Aventa': {'SMAPE': 13.076, 'ICR': 86.492, 'IS': 3.244},
        'T1': {'SMAPE': 5.321, 'ICR': 84.394, 'IS': 5.466}
    },
    'Regional Model': {
        'Site1': {'SMAPE': 2.709, 'ICR': 90.737, 'IS': 3.600},
        'Site2': {'SMAPE': 2.709, 'ICR': 91.263, 'IS': 3.800},
        'Site3': {'SMAPE': 2.789, 'ICR': 88.815, 'IS': 3.612},
        'Aventa': {'SMAPE': 11.849, 'ICR': 87.103, 'IS': 2.839},
        'T1': {'SMAPE': 4.552, 'ICR': 85.305, 'IS': 5.310}
    }
}

# 转换为适合绘图的格式
sites = ['Site1', 'Site2', 'Site3', 'Aventa', 'T1']
metrics = ['SMAPE', 'ICR', 'IS']

# 创建图形
fig, axes = plt.subplots(2, 2, figsize=(16, 14))
fig.suptitle('Model Performance Comparison: Global vs Regional Models', 
             fontsize=20, fontweight='bold', y=1.02)

# 设置颜色方案
model_colors = {
    'Global Model': '#1f77b4',  # 蓝色
    'Regional Model': '#ff7f0e'  # 橙色
}

site_colors = cm.Set2(np.linspace(0, 1, len(sites)))

# 子图1：综合气泡图
ax1 = axes[0, 0]

# 绘制气泡图
for i, site in enumerate(sites):
    for model_name, model_data in data.items():
        # 获取数据
        smape = model_data[site]['SMAPE']
        icr = model_data[site]['ICR']
        is_val = model_data[site]['IS']
        
        # 调整气泡大小（IS值越小，气泡越大 - 因为IS越小越好）
        # 我们取倒数，使小的IS值对应大的气泡
        bubble_size = 1/is_val * 300  # 乘以系数调整大小
        
        # 绘制气泡
        ax1.scatter(smape, icr, s=bubble_size, 
                   alpha=0.7, 
                   color=site_colors[i] if model_name == 'Global Model' else model_colors[model_name],
                   edgecolor='white', linewidth=1.5,
                   label=f'{site} ({model_name})' if i == 0 and model_name == 'Global Model' else '',
                   marker='o' if model_name == 'Global Model' else 's')
        
        # 添加站点标签
        ax1.annotate(site, (smape, icr), 
                    fontsize=10, ha='center', va='center',
                    color='black', fontweight='bold')

ax1.set_xlabel('SMAPE (越小越好)', fontsize=14)
ax1.set_ylabel('ICR (%) (越大越好)', fontsize=14)
ax1.set_title('(a) 综合性能气泡图\n气泡大小∝1/IS (IS越小气泡越大)', fontsize=16, fontweight='bold', pad=15)
ax1.grid(True, alpha=0.3)

# 添加图例
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markersize=10, label='Global Model'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='gray', markersize=10, label='Regional Model'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=site_colors[0], markersize=10, label='Site1'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=site_colors[1], markersize=10, label='Site2'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=site_colors[2], markersize=10, label='Site3'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=site_colors[3], markersize=10, label='Aventa'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=site_colors[4], markersize=10, label='T1'),
]
ax1.legend(handles=legend_elements, loc='upper left', fontsize=10, framealpha=0.9)

# 子图2：按模型分组的雷达图样式气泡图
ax2 = axes[0, 1]

# 为每个模型创建子图
for model_idx, (model_name, model_data) in enumerate(data.items()):
    # 计算每个站点的平均位置
    angles = np.linspace(0, 2*np.pi, len(sites), endpoint=False).tolist()
    angles += angles[:1]  # 闭合
    
    # 计算归一化的指标值
    smape_vals = [model_data[site]['SMAPE'] for site in sites]
    icr_vals = [model_data[site]['ICR'] for site in sites]
    is_vals = [model_data[site]['IS'] for site in sites]
    
    # 归一化处理（使所有指标在0-1范围内，注意SMAPE和IS越小越好，ICR越大越好）
    smape_norm = 1 - (np.array(smape_vals) - min(smape_vals)) / (max(smape_vals) - min(smape_vals))
    icr_norm = (np.array(icr_vals) - min(icr_vals)) / (max(icr_vals) - min(icr_vals))
    is_norm = 1 - (np.array(is_vals) - min(is_vals)) / (max(is_vals) - min(is_vals))
    
    # 计算综合得分
    composite_score = (smape_norm + icr_norm + is_norm) / 3
    
    # 转换为雷达图格式
    smape_norm = np.concatenate((smape_norm, [smape_norm[0]]))
    icr_norm = np.concatenate((icr_norm, [icr_norm[0]]))
    composite_score = np.concatenate((composite_score, [composite_score[0]]))
    
    # 绘制雷达图样式的气泡
    offset = -0.1 if model_name == 'Global Model' else 0.1
    for i, site in enumerate(sites):
        # 计算气泡大小（基于综合得分）
        bubble_size = composite_score[i] * 500 + 100
        
        # 绘制气泡
        ax2.scatter(angles[i] + offset, composite_score[i], 
                   s=bubble_size,
                   alpha=0.7,
                   color=site_colors[i],
                   edgecolor='white', linewidth=1.5,
                   label=site if model_idx == 0 and i < len(sites) else '',
                   marker='o' if model_name == 'Global Model' else 's')

# 设置极坐标样式
ax2.set_xticks(angles[:-1])
ax2.set_xticklabels(sites, fontsize=12)
ax2.set_ylim(0, 1)
ax2.set_ylabel('归一化综合得分', fontsize=12)
ax2.set_title('(b) 雷达图样式气泡图\n气泡大小∝综合性能', fontsize=16, fontweight='bold', pad=15)

# 子图3：SMAPE vs ICR 详细对比
ax3 = axes[1, 0]

# 绘制每个站点的两个模型对比
for i, site in enumerate(sites):
    # 获取数据
    g_smape = data['Global Model'][site]['SMAPE']
    g_icr = data['Global Model'][site]['ICR']
    r_smape = data['Regional Model'][site]['SMAPE']
    r_icr = data['Regional Model'][site]['ICR']
    
    # 绘制连接线
    ax3.plot([g_smape, r_smape], [g_icr, r_icr], 
             color=site_colors[i], alpha=0.5, linewidth=1.5, zorder=1)
    
    # 绘制气泡
    ax3.scatter(g_smape, g_icr, s=300, 
               alpha=0.8, color=site_colors[i], 
               edgecolor='white', linewidth=2,
               marker='o', label=f'{site} (Global)' if i == 0 else '')
    
    ax3.scatter(r_smape, r_icr, s=300, 
               alpha=0.8, color=site_colors[i], 
               edgecolor='white', linewidth=2,
               marker='s', label=f'{site} (Regional)' if i == 0 else '')
    
    # 添加站点标签
    ax3.annotate(f'G\n{site}', (g_smape, g_icr), 
                fontsize=10, ha='center', va='center',
                color='white', fontweight='bold')
    
    ax3.annotate(f'R\n{site}', (r_smape, r_icr), 
                fontsize=10, ha='center', va='center',
                color='white', fontweight='bold')

ax3.set_xlabel('SMAPE (越小越好)', fontsize=14)
ax3.set_ylabel('ICR (%) (越大越好)', fontsize=14)
ax3.set_title('(c) 站点详细对比\n连接线显示模型间差异', fontsize=16, fontweight='bold', pad=15)
ax3.grid(True, alpha=0.3)

# 添加性能改善箭头
for i, site in enumerate(sites):
    g_smape = data['Global Model'][site]['SMAPE']
    g_icr = data['Global Model'][site]['ICR']
    r_smape = data['Regional Model'][site]['SMAPE']
    r_icr = data['Regional Model'][site]['ICR']
    
    # 计算改善程度
    smape_improve = g_smape - r_smape  # 正数表示改善
    icr_improve = r_icr - g_icr  # 正数表示改善
    
    # 只在有显著改善时添加箭头
    if smape_improve > 0.1 or icr_improve > 0.1:
        ax3.annotate('', xy=(r_smape, r_icr), xytext=(g_smape, g_icr),
                    arrowprops=dict(arrowstyle='->', color=site_colors[i], 
                                   linewidth=2, alpha=0.7))

# 子图4：性能改善热力图气泡
ax4 = axes[1, 1]

# 计算每个指标的性能改善
improvements = {}
for site in sites:
    improvements[site] = {
        'SMAPE_improve': data['Global Model'][site]['SMAPE'] - data['Regional Model'][site]['SMAPE'],
        'ICR_improve': data['Regional Model'][site]['ICR'] - data['Global Model'][site]['ICR'],
        'IS_improve': data['Global Model'][site]['IS'] - data['Regional Model'][site]['IS']
    }

# 创建热力图样式气泡图
x_pos = np.arange(len(sites))
metrics_improve = ['SMAPE_improve', 'ICR_improve', 'IS_improve']
colors_improve = ['#2ca02c', '#d62728', '#9467bd']  # 绿色: SMAPE改善, 红色: ICR改善, 紫色: IS改善

# 设置条形图位置
bar_width = 0.25
for i, metric in enumerate(metrics_improve):
    values = [improvements[site][metric] for site in sites]
    
    # 绘制气泡（使用条形图位置但用气泡表示）
    for j, (site, value) in enumerate(zip(sites, values)):
        # 气泡大小基于改善程度绝对值
        bubble_size = abs(value) * 2000
        
        # 绘制气泡
        ax4.scatter(j + i*bar_width, value, 
                   s=bubble_size,
                   alpha=0.7,
                   color=colors_improve[i],
                   edgecolor='white', linewidth=1.5,
                   label=metric.replace('_improve', '') if j == 0 else '')

# 设置x轴
ax4.set_xticks(x_pos + bar_width)
ax4.set_xticklabels(sites, fontsize=12)
ax4.set_xlabel('站点', fontsize=14)
ax4.set_ylabel('性能改善值', fontsize=14)
ax4.set_title('(d) 性能改善热力图气泡\n正值表示Regional Model更优', fontsize=16, fontweight='bold', pad=15)
ax4.grid(True, alpha=0.3)
ax4.axhline(y=0, color='black', linewidth=1, linestyle='-', alpha=0.5)

# 添加图例
improve_legend = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor=colors_improve[0], markersize=10, label='SMAPE改善'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=colors_improve[1], markersize=10, label='ICR改善'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor=colors_improve[2], markersize=10, label='IS改善')
]
ax4.legend(handles=improve_legend, loc='upper right', fontsize=10, framealpha=0.9)

# 调整布局
plt.tight_layout()

# 添加总结性文字
summary_text = """性能总结:
1. Regional Model在所有站点的SMAPE都更低（更好）
2. Regional Model在大多数站点的ICR更高（更好）
3. Regional Model在IS指标上表现不一
4. Aventa站点在两个模型上表现都相对较差"""
fig.text(0.02, 0.02, summary_text, fontsize=11, 
         verticalalignment='bottom',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='gray'))

plt.show()
```
